# Task
Analyze the 'Samsung_mobile.csv' dataset by loading it into a DataFrame, exploring its structure, filtering based on relevant conditions, grouping and aggregating the data to calculate statistics, visualizing the aggregated data with legends, and summarizing the key findings.

## Load Samsung Mobile Data

### Subtask:
Load the 'Samsung_mobile.csv' file into a pandas DataFrame for analysis.


**Reasoning**:
To load the 'Samsung_mobile.csv' file into a pandas DataFrame, I will first import the pandas library and then use `pd.read_csv()` to load the data into `df_samsung`. Finally, I will display the head of the DataFrame to confirm successful loading.



```python
import pandas as pd
df_samsung = pd.read_csv('/content/Samsung_mobile.csv')
print("Samsung mobile data loaded successfully.")
df_samsung.head()
```

## Explore Samsung Data

### Subtask:
Perform initial data exploration on the `Samsung_mobile` DataFrame, including displaying the first few rows, checking column information, and reviewing descriptive statistics to understand its structure and identify relevant columns for subsequent steps.


**Reasoning**:
To perform initial data exploration as requested, I will display the first few rows, show column information, and generate descriptive statistics of the `df_samsung` DataFrame in a single code block.



```python
print("First 5 rows of df_samsung:")
print(df_samsung.head())

print("\nDataFrame Info:")
df_samsung.info()

print("\nDescriptive Statistics:")
print(df_samsung.describe())
```

## Filter Samsung Data

### Subtask:
Apply filtering techniques to the `Samsung_mobile` DataFrame based on appropriate conditions from its columns (e.g., filtering by 'Mobile_Phone' type, 'Review' status, or other relevant attributes) to focus on a specific subset of the data.


**Reasoning**:
To filter the `df_samsung` DataFrame as per the instructions, I will apply two conditions: 'Review' equals 'good' and 'RAM_GB' is greater than or equal to 6. The result will be stored in `df_samsung_filtered`, and its head and shape will be displayed for verification.



```python
df_samsung_filtered = df_samsung[(df_samsung['Review'] == 'good') & (df_samsung['RAM_GB'] >= 6)]

print("First five rows of the filtered Samsung DataFrame (df_samsung_filtered):")
print(df_samsung_filtered.head())

print(f"\nShape of the filtered Samsung DataFrame (df_samsung_filtered): {df_samsung_filtered.shape}")
```

## Group and Aggregate Samsung Data

### Subtask:
Group the filtered Samsung data by relevant categorical variables (e.g., 'Model', 'RAM_GB') and compute aggregate statistics such as sum, mean, count, minimum, and maximum for appropriate numerical columns like 'Latest_Price_Taka' or 'Discount'.


**Reasoning**:
To group the filtered Samsung data as instructed, I will use the `groupby()` method twice, once for 'Model' and once for 'RAM_GB', applying multiple aggregation functions (mean, sum, count, min, max) to the specified numerical columns. Afterwards, I will display the head of both resulting dataframes to confirm the aggregation.



```python
df_samsung_grouped_model = df_samsung_filtered.groupby('Model').agg(
    Latest_Price_Taka_Mean=('Latest_Price_Taka', 'mean'),
    Latest_Price_Taka_Sum=('Latest_Price_Taka', 'sum'),
    Latest_Price_Taka_Count=('Latest_Price_Taka', 'count'),
    Latest_Price_Taka_Min=('Latest_Price_Taka', 'min'),
    Latest_Price_Taka_Max=('Latest_Price_Taka', 'max'),
    Discount_Mean=('Discount', 'mean'),
    Discount_Sum=('Discount', 'sum'),
    Discount_Count=('Discount', 'count'),
    Discount_Min=('Discount', 'min'),
    Discount_Max=('Discount', 'max'),
    Selling_Price_Mean=('Selling_Price', 'mean'),
    Selling_Price_Sum=('Selling_Price', 'sum'),
    Selling_Price_Count=('Selling_Price', 'count'),
    Selling_Price_Min=('Selling_Price', 'min'),
    Selling_Price_Max=('Selling_Price', 'max')
)

df_samsung_grouped_ram = df_samsung_filtered.groupby('RAM_GB').agg(
    Latest_Price_Taka_Mean=('Latest_Price_Taka', 'mean'),
    Latest_Price_Taka_Sum=('Latest_Price_Taka', 'sum'),
    Latest_Price_Taka_Count=('Latest_Price_Taka', 'count'),
    Latest_Price_Taka_Min=('Latest_Price_Taka', 'min'),
    Latest_Price_Taka_Max=('Latest_Price_Taka', 'max'),
    Discount_Mean=('Discount', 'mean'),
    Discount_Sum=('Discount', 'sum'),
    Discount_Count=('Discount', 'count'),
    Discount_Min=('Discount', 'min'),
    Discount_Max=('Discount', 'max'),
    Selling_Price_Mean=('Selling_Price', 'mean'),
    Selling_Price_Sum=('Selling_Price', 'sum'),
    Selling_Price_Count=('Selling_Price', 'count'),
    Selling_Price_Min=('Selling_Price', 'min'),
    Selling_Price_Max=('Selling_Price', 'max')
)

print("First five rows of grouped data by Model:\n")
print(df_samsung_grouped_model.head())

print("\nFirst five rows of grouped data by RAM_GB:\n")
print(df_samsung_grouped_ram.head())
```

## Visualize Aggregated Samsung Data

### Subtask:
Generate a plot (e.g., a bar chart or line plot) to visualize the aggregated Samsung data, providing insights into trends or comparisons. The plot will include legends for clarity.


**Reasoning**:
To visualize the aggregated data, I will import `matplotlib.pyplot` and `seaborn`, and then create a bar plot comparing 'Latest_Price_Taka_Mean' and 'Selling_Price_Mean' by 'RAM_GB', ensuring proper labels, title, and a legend for clarity.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reset index to make 'RAM_GB' a column for plotting with seaborn
df_plot = df_samsung_grouped_ram.reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(
x='RAM_GB',
y='Latest_Price_Taka_Mean',
data=df_plot,
color='skyblue',
label='Latest Price'
)
sns.barplot(
x='RAM_GB',
y='Selling_Price_Mean',
data=df_plot,
color='lightcoral',
label='Selling Price'
)

plt.title('Average Latest Price and Selling Price by RAM (GB)')
plt.xlabel('RAM (GB)')
plt.ylabel('Average Price (Taka)')
plt.legend()
plt.show()
```
