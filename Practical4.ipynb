1. Generate a normal distribution and plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
print("Libraries imported: numpy and matplotlib.pyplot.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers (mean = 0, std = 1)
data = np.random.normal(loc=0, scale=1, size=1000)

# Print first 10 values as sample
print("Sample values:", data[:10])

# Plot histogram
plt.hist(data, bins=30)
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

2. Compute mean, median, variance, and standard deviation for a dataset.

```python
import numpy as np # Keep numpy for potential future use or if user reconsiders

# Take input from user
data = list(map(int, input("Enter numbers separated by space: ").split()))

# ----- Mean -----
mean = sum(data) / len(data)

# ----- Median -----
data.sort()
n = len(data)

if n % 2 == 0:
    median = (data[n//2 - 1] + data[n//2]) / 2
else:
    median = data[n//2]

# ----- Mode -----
# Note: If there are multiple modes, this will return the first one encountered
from collections import Counter
counts = Counter(data)
max_count = max(counts.values())
modes = [key for key, value in counts.items() if value == max_count]

# ----- Variance (using for loop) -----
sum_sq_diff = 0
for x in data:
    sum_sq_diff += (x - mean) ** 2
variance = sum_sq_diff / n

# ----- Standard Deviation-----
std_dev = variance ** 0.5


print("Mean =", mean)
print("Median =", median)
print("Mode(s) =", modes)
print("Variance =", variance)
print("Standard Deviation =", std_dev)
```

3. Simulate a coin toss experiment and compute probabilities.

```python
num_tosses = 1000
coin_toss_results = np.random.choice(['Heads', 'Tails'], size=num_tosses)

print("Simulating {num_tosses} coin tosses.")
print("First 10 results:", coin_toss_results[:10])
print("Coin toss simulation completed.")
```

```python
unique_outcomes, counts = np.unique(coin_toss_results, return_counts=True)
probabilities = counts / num_tosses

for outcome, prob in zip(unique_outcomes, probabilities):
    print(f"Observed probability of {outcome}: {prob:.4f}")

print("Observed probabilities calculated and displayed.")
```

```python
plt.bar(unique_outcomes, probabilities, color=['skyblue', 'lightcoral'])
plt.title('Observed Probabilities of Coin Toss Outcomes')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.ylim(0, 1) # Probabilities are between 0 and 1
plt.show()
print("Observed probabilities visualized.")
```
