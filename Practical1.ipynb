## Compute Factorial



```python
def factorial(n):
    if n < 0:
        return "Factorial does not exist for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

print("\n--- Factorial Calculator ---")
num_factorial = int(input("Enter a number to find its factorial: "))
print("Factorial of", num_factorial, "is", factorial(num_factorial))
```

## Display Multiplication Table




```python
print("\n--- Multiplication Table ---")
num_table = int(input("Enter a number to display its multiplication table: "))
print(f"Multiplication Table for {num_table}:")
for i in range(1, 11):
    print(f"{num_table} x {i} = {num_table * i}")

```

## Identify Odd or Even




```python
print("\n--- Odd or Even Check ---")
num_odd_even = int(input("Enter a number to check if it's odd or even: "))
if num_odd_even % 2 == 0:
    print(f"{num_odd_even} is an EVEN number.")
else:
    print(f"{num_odd_even} is an ODD number.")
```

## Check Palindrome Number



```python
print("\n--- Palindrome Check ---")
num_palindrome = input("Enter a number (or word) to check if it's a palindrome: ")
if str(num_palindrome) == str(num_palindrome)[::-1]:
    print(f"{num_palindrome} is a PALINDROME.")
else:
    print(f"{num_palindrome} is NOT a PALINDROME.")

```

## Develop Simple Expert System



```python
print("\n--- Simple Expert System (Computer Troubleshooting) ---")
print("Let's diagnose your computer problem.")

question1 = input("Is your computer completely unresponsive (no power, no display)? (yes/no): ").lower()

if question1 == 'yes':
    question2 = input("Is the power cable securely plugged in? (yes/no): ").lower()
    if question2 == 'yes':
        question3 = input("Have you tried a different power outlet or checked the power supply? (yes/no): ").lower()
        if question3 == 'yes':
            print("Recommendation: Your power supply or motherboard might be faulty. Consider professional repair.")
        else:
            print("Recommendation: Try plugging into a different power outlet or checking your power supply.")
    else:
        print("Recommendation: Please ensure all power cables are securely plugged into the computer and the wall.")
else:
    question4 = input("Is your computer turning on but showing no display? (yes/no): ").lower()
    if question4 == 'yes':
        question5 = input("Are your monitor cables securely connected? (yes/no): ").lower()
        if question5 == 'yes':
            print("Recommendation: Try reseating your RAM or graphics card. If the issue persists, your graphics card might be faulty.")
        else:
            print("Recommendation: Ensure your monitor cables are securely plugged into both the monitor and the computer.")
    else:
        question6 = input("Is your computer generally slow or freezing often? (yes/no): ").lower()
        if question6 == 'yes':
            print("Recommendation: Check for malware, clear temporary files, or consider upgrading RAM/storage. A fresh OS install might also help.")
        else:
            print("Recommendation: It seems your computer is working fine! If you have another issue, please describe it.")

print("--- End of Expert System ---")
```
