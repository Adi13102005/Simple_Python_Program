## Compute Factorial

### Subtask:
Write a Python function to compute the factorial of a given number. Include examples to demonstrate its usage.


**Reasoning**:
The subtask requires writing a Python function to compute the factorial, including handling edge cases and demonstrating its usage. This code block will define the function, implement the required logic, error handling, and test cases.



```python
def compute_factorial(n):
    if not isinstance(n, int) or n < 0:
        raise ValueError("Input must be a non-negative integer.")

    if n == 0:
        return 1

    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    return factorial

# Demonstrate usage with examples
print(f"Factorial of 0: {compute_factorial(0)}")
print(f"Factorial of 1: {compute_factorial(1)}")
print(f"Factorial of 5: {compute_factorial(5)}")

# Test with invalid inputs
try:
    print(f"Factorial of -3: {compute_factorial(-3)}")
except ValueError as e:
    print(f"Error: {e}")

try:
    print(f"Factorial of 3.5: {compute_factorial(3.5)}")
except ValueError as e:
    print(f"Error: {e}")
```

## Display Multiplication Table

### Subtask:
Write a Python function that takes a number as input and displays its multiplication table.


**Reasoning**:
The subtask requires a Python function to display the multiplication table for a given number. This code block will define the `display_multiplication_table` function, implement the multiplication logic using a loop, ensure input validation for a positive integer, and demonstrate its usage with an example.

**Reasoning**:
Next, I will implement the `display_multiplication_table` function according to the instructions, including input validation, the multiplication loop, formatted output, and an example call.



```python
def display_multiplication_table(number):
    if not isinstance(number, int) or number <= 0:
        raise ValueError("Input must be a positive integer.")

    print(f"\nMultiplication Table for {number}:")
    for i in range(1, 11): # Iterate from 1 to 10
        product = number * i
        print(f"{number} x {i} = {product}")

# Demonstrate usage with an example number
try:
    display_multiplication_table(5)
except ValueError as e:
    print(f"Error: {e}")

# Test with invalid inputs
try:
    display_multiplication_table(-3)
except ValueError as e:
    print(f"Error: {e}")

try:
    display_multiplication_table(3.5)
except ValueError as e:
    print(f"Error: {e}")
```

## Identify Odd or Even

### Subtask:
Write a Python function to determine if a given number is odd or even. Provide test cases to verify the function.


A Python function `is_odd_or_even` will be defined to determine if a given number is odd or even. The function will include input validation to ensure the input is an integer. It will use the modulo operator to check for divisibility by 2 and return 'Even' or 'Odd'. Finally, the code block will include comprehensive test cases for valid and invalid inputs, demonstrating the function's correct behavior and error handling.

**Reasoning**:
Now, I will implement the `is_odd_or_even` function as described, including input validation, the core logic for checking odd/even, and comprehensive test cases to demonstrate its usage and error handling.



```python
def is_odd_or_even(number):
    if not isinstance(number, int):
        raise ValueError("Input must be an integer.")

    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Demonstrate usage with test cases
print(f"0 is {is_odd_or_even(0)}")
print(f"2 is {is_odd_or_even(2)}")
print(f"5 is {is_odd_or_even(5)}")
print(f"-4 is {is_odd_or_even(-4)}")
print(f"-7 is {is_odd_or_even(-7)}")

# Test with invalid inputs
try:
    print(f"3.5 is {is_odd_or_even(3.5)}")
except ValueError as e:
    print(f"Error: {e}")

try:
    print(f"'hello' is {is_odd_or_even('hello')}")
except ValueError as e:
    print(f"Error: {e}")
```

## Check Palindrome Number

### Subtask:
Write a Python function to check if an entered number is a palindrome (reads the same forwards and backward).


A Python function `is_palindrome_number` will be defined to check if a given integer is a palindrome. The function will first validate that the input is an integer. Then, the number will be converted to a string to facilitate comparison with its reversed version. Finally, test cases will be provided to demonstrate the function's behavior with both palindrome and non-palindrome numbers, as well as handling invalid inputs.

**Reasoning**:
Now, I will implement the `is_palindrome_number` function as described, including input validation, the core logic for checking palindromes, and comprehensive test cases to demonstrate its usage and error handling.



```python
def is_palindrome_number(number):
    if not isinstance(number, int):
        raise ValueError("Input must be an integer.")

    # Convert the integer to a string
    num_str = str(number)

    # Compare the original string with its reversed version
    return num_str == num_str[::-1]

# Demonstrate usage with examples
print(f"Is 121 a palindrome? {is_palindrome_number(121)}")
print(f"Is 12321 a palindrome? {is_palindrome_number(12321)}")
print(f"Is 123 a palindrome? {is_palindrome_number(123)}")
print(f"Is -121 a palindrome? {is_palindrome_number(-121)}") # -121 is not considered a palindrome in string form
print(f"Is 0 a palindrome? {is_palindrome_number(0)}")
print(f"Is 1 a palindrome? {is_palindrome_number(1)}")
print(f"Is 10 a palindrome? {is_palindrome_number(10)}")

# Test with invalid inputs
try:
    print(f"Is 12.1 a palindrome? {is_palindrome_number(12.1)}")
except ValueError as e:
    print(f"Error: {e}")

try:
    print(f"Is 'hello' a palindrome? {is_palindrome_number('hello')}")
except ValueError as e:
    print(f"Error: {e}")
```

## Develop Simple Expert System

### Subtask:
Develop a simple rule-based expert system in Python that provides decision-making based on a set of predefined rules. The system should be able to take user input and give a relevant output based on its knowledge base.


A Python function `simple_expert_system` will be defined to simulate a rule-based expert system. This function will encapsulate a dictionary representing a knowledge base, mapping keywords or conditions to predefined advice. It will prompt the user for input, convert it to lowercase for case-insensitive matching, and then iterate through the knowledge base to find matching keywords. If a match is found, the corresponding advice will be printed. A default response will be provided if no rules are matched. Finally, several example calls will demonstrate the system's ability to respond to various user queries, including those that match rules and those that do not.

**Reasoning**:
Now, I will implement the `simple_expert_system` function according to the detailed instructions. This will involve defining the function, creating a knowledge base, handling user input, applying rule-based logic, and providing example calls to demonstrate its functionality.



```python
def simple_expert_system(user_query):
    # 2. Define a knowledge base as a dictionary
    knowledge_base = {
        "problem with internet": "Try restarting your router and modem. If that doesn't work, contact your internet service provider.",
        "computer slow": "Close unnecessary applications and browser tabs. Consider running a disk cleanup or malware scan.",
        "email not sending": "Check your internet connection and ensure your email client settings (SMTP server, port, authentication) are correct.",
        "printer not working": "Check if the printer is turned on and connected. Ensure ink/toner and paper levels are adequate. Try reinstalling drivers.",
        "software not installing": "Check system requirements, free disk space, and administrator permissions. Disable antivirus temporarily if necessary."
    }

    # 4. Process the user's input (convert to lowercase)
    query_lower = user_query.lower()

    # 5. Implement the rule-based decision-making logic
    found_advice = False
    for keyword, advice in knowledge_base.items():
        if keyword in query_lower:
            print(f"Expert System Advice: {advice}")
            found_advice = True
            break

    # 7. Include a default response
    if not found_advice:
        print("Expert System: I'm sorry, I don't have specific advice for that. Please try rephrasing your query or contact support.")

# 8. Provide example calls to your simple_expert_system function
print("--- Expert System Demonstrations ---")
simple_expert_system("I have a problem with my internet connection.")
simple_expert_system("My computer is running very slow.")
simple_expert_system("Why is my EMAIL not sending?")
simple_expert_system("My printer isn't working.")
simple_expert_system("I can't install new software.")
simple_expert_system("My coffee machine is broken.")
simple_expert_system("What's the weather like today?")

print("------------------------------------")
```
