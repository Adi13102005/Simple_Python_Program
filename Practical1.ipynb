## Compute Factorial



```python
def compute_factorial(n):
    if not isinstance(n, int) or n < 0:
        raise ValueError("Input must be a non-negative integer.")

    if n == 0:
        return 1

    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    return factorial

# Demonstrate usage with examples
print(f"Factorial of 0: {compute_factorial(0)}")
print(f"Factorial of 1: {compute_factorial(1)}")
print(f"Factorial of 5: {compute_factorial(5)}")

# Test with invalid inputs
try:
    print(f"Factorial of -3: {compute_factorial(-3)}")
except ValueError as e:
    print(f"Error: {e}")

try:
    print(f"Factorial of 3.5: {compute_factorial(3.5)}")
except ValueError as e:
    print(f"Error: {e}")
```

## Display Multiplication Table




```python
def display_multiplication_table(number):
    if not isinstance(number, int) or number <= 0:
        raise ValueError("Input must be a positive integer.")

    print(f"\nMultiplication Table for {number}:")
    for i in range(1, 11): # Iterate from 1 to 10
        product = number * i
        print(f"{number} x {i} = {product}")

# Demonstrate usage with an example number
try:
    display_multiplication_table(5)
except ValueError as e:
    print(f"Error: {e}")

# Test with invalid inputs
try:
    display_multiplication_table(-3)
except ValueError as e:
    print(f"Error: {e}")

try:
    display_multiplication_table(3.5)
except ValueError as e:
    print(f"Error: {e}")
```

## Identify Odd or Even




```python
def is_odd_or_even(number):
    if not isinstance(number, int):
        raise ValueError("Input must be an integer.")

    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Demonstrate usage with test cases
print(f"0 is {is_odd_or_even(0)}")
print(f"2 is {is_odd_or_even(2)}")
print(f"5 is {is_odd_or_even(5)}")
print(f"-4 is {is_odd_or_even(-4)}")
print(f"-7 is {is_odd_or_even(-7)}")

# Test with invalid inputs
try:
    print(f"3.5 is {is_odd_or_even(3.5)}")
except ValueError as e:
    print(f"Error: {e}")

try:
    print(f"'hello' is {is_odd_or_even('hello')}")
except ValueError as e:
    print(f"Error: {e}")
```

## Check Palindrome Number



```python
def is_palindrome_number(number):
    if not isinstance(number, int):
        raise ValueError("Input must be an integer.")

    # Convert the integer to a string
    num_str = str(number)

    # Compare the original string with its reversed version
    return num_str == num_str[::-1]

# Demonstrate usage with examples
print(f"Is 121 a palindrome? {is_palindrome_number(121)}")
print(f"Is 12321 a palindrome? {is_palindrome_number(12321)}")
print(f"Is 123 a palindrome? {is_palindrome_number(123)}")
print(f"Is -121 a palindrome? {is_palindrome_number(-121)}") # -121 is not considered a palindrome in string form
print(f"Is 0 a palindrome? {is_palindrome_number(0)}")
print(f"Is 1 a palindrome? {is_palindrome_number(1)}")
print(f"Is 10 a palindrome? {is_palindrome_number(10)}")

# Test with invalid inputs
try:
    print(f"Is 12.1 a palindrome? {is_palindrome_number(12.1)}")
except ValueError as e:
    print(f"Error: {e}")

try:
    print(f"Is 'hello' a palindrome? {is_palindrome_number('hello')}")
except ValueError as e:
    print(f"Error: {e}")
```
