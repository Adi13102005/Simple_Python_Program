1. Breadth-First Search (BFS) and Depth-First Search (DFS)

```python
# Graph representation
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

# Breadth First Search
def bfs(graph, start):
    visited = []
    queue = [start]

    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.append(node)
            queue.extend(graph[node])

    return visited




print("BFS Traversal:", bfs(graph, 'A'))

```

```python
# Graph representation using dictionary
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

# Depth First Search function
def dfs(graph, start, visited=None):
    if visited is None:
        visited = []

    visited.append(start)
    print(start, end=" ")

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)


print("DFS Traversal:")
dfs(graph, 'A')
```

2. A* Algorithm

```python
# Graph with cost
graph = {
    'A': [('B', 1), ('C', 3)],
    'B': [('D', 3)],
    'C': [('D', 1)],
    'D': []
}

# Heuristic values
heuristic = {
    'A': 4,
    'B': 2,
    'C': 2,
    'D': 0
}

def a_star(start, goal):
    open_list = [(heuristic[start], start, 0)]
    visited = []

    while open_list:
        open_list.sort()
        f, node, cost = open_list.pop(0)

        if node == goal:
            return cost

        visited.append(node)

        for neighbor, weight in graph[node]:
            if neighbor not in visited:
                g = cost + weight
                h = heuristic[neighbor]
                open_list.append((g + h, neighbor, g))

    return None


print("Shortest path cost using A*:", a_star('A', 'D'))
```

3. MiniMax Algorithm

```python
# Minimax implementation
def minimax(depth, node_index, is_max, scores, height):

    if depth == height:
        return scores[node_index]

    if is_max:
        return max(
            minimax(depth + 1, node_index * 2, False, scores, height),
            minimax(depth + 1, node_index * 2 + 1, False, scores, height)
        )
    else:
        return min(
            minimax(depth + 1, node_index * 2, True, scores, height),
            minimax(depth + 1, node_index * 2 + 1, True, scores, height)
        )


# Leaf node scores
scores = [6, 5, 2, 9]

import math
height = int(math.log2(len(scores)))

print("Optimal value using Minimax:", minimax(0, 0, True, scores, height))
```
